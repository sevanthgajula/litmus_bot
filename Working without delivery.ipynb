```python
import telegram
from telegram.ext import Updater, CommandHandler, MessageHandler, Filters

# define the inventory
inventory = [    {'name': 'Soap', 'price': 20, 'quantity': 50},    
                 {'name': 'Toothpaste', 'price': 30, 'quantity': 30},    
                 {'name': 'Shampoo', 'price': 50, 'quantity': 20},    
                 {'name': 'Cleanser', 'price': 10, 'quantity': 100},    
                 {'name': 'Conditioner', 'price': 60, 'quantity': 15},    
                 {'name': 'Hand Sanitizer', 'price': 25, 'quantity': 40},    
                 {'name': 'Face Wash', 'price': 35, 'quantity': 25},    
                 {'name': 'Mouthwash', 'price': 15, 'quantity': 50},    
                 {'name': 'Razor Blades', 'price': 12, 'quantity': 75},    
                 {'name': 'Hair Gel', 'price': 20, 'quantity': 30},    
                 {'name': 'Deodorant', 'price': 18, 'quantity': 40},    
                 {'name': 'Sunscreen', 'price': 22, 'quantity': 30}]


# define a dictionary to store the user's cart
cart = {}

# define the command handlers
def start(update, context):
    """Send a message when the command /start is issued."""
    update.message.reply_text('Welcome to the General Store! Type /help to see the available commands.')

def help(update, context):
    """Send a message when the command /help is issued."""
    help_message = '''
    Available commands:
    /list - Show the list of available items
    /cart - Show the items in your cart
    /add <item name> <quantity> - Add an item to your cart
    /remove <item name> - Remove an item from your cart
    /confirm - Confirm your order
    '''
    update.message.reply_text(help_message)

def list_items(update, context):
    """Send a message with the list of available items."""
    item_list = '\n'.join([f"{item['name']} ({item['quantity']} left) - Rs. {item['price']}" for item in inventory])
    update.message.reply_text(f"The following items are available for purchase:\n{item_list}")

def add_item(update, context):
    """Add an item to the user's cart."""
    message = update.message.text
    item_name, item_quantity = message.split()[1:]
    item_quantity = int(item_quantity)
    for item in inventory:
        if item['name'].lower() == item_name.lower():
            if item_quantity <= item['quantity']:
                if item_name in cart:
                    cart[item_name] += item_quantity
                else:
                    cart[item_name] = item_quantity
                item['quantity'] -= item_quantity
                update.message.reply_text(f"{item_quantity} {item_name}(s) added to your cart.")
            else:
                update.message.reply_text(f"Sorry, we only have {item['quantity']} {item_name}(s) left.")
            break
    else:
        update.message.reply_text(f"Sorry, we do not have {item_name}.")

def remove_item(update, context):
    """Remove an item from the user's cart."""
    message = update.message.text
    item_name = message.split()[1]
    if item_name in cart:
        for item in inventory:
            if item['name'].lower() == item_name.lower():
                item['quantity'] += cart[item_name]
                break
        del cart[item_name]
        update.message.reply_text(f"{item_name} removed from your cart.")
    else:
        update.message.reply_text(f"You do not have {item_name} in your cart.")

def show_cart(update, context):
    """Show the items in the user's cart."""
    if cart:
        cart_list = '\n'.join([f"{item} - {cart[item]}" for item in cart])
        update.message.reply_text(f"Your cart contains:\n{cart_list}")
    else:
        update.message.reply_text("Your cart is empty.")

def confirm_order(update, context):
    """Confirm the user's order."""
    if cart:
        total_cost = sum([inventory[i]['price'] * cart[item] for i, item in enumerate(cart)])
        confirmation_message = f"Your order of Rs. {total_cost} has been confirmed. The following items will be delivered:\n"
        for item in cart:
            confirmation_message += f"{item} - {cart[item]}\n"
        update.message.reply_text(confirmation_message)
        owner_message = f"New order received:\n{confirmation_message}"
        context.bot.send_message(chat_id='@store_owner', text=owner_message)
        cart.clear()
    else:
        update.message.reply_text("Your cart is empty. Add some items to confirm your order.")

# create the Telegram bot and add the handlers
TOKEN = '5673837379:AAF1A5segzctmk4uP33dwXQKkn-GVrNujmE'
bot = telegram.Bot(token=TOKEN)
updater = Updater(token=TOKEN, use_context=True)
dispatcher = updater.dispatcher

# define the handlers
dispatcher.add_handler(CommandHandler('start', start))
dispatcher.add_handler(CommandHandler('help', help))
dispatcher.add_handler(CommandHandler('list', list_items))
dispatcher.add_handler(CommandHandler('cart', show_cart))
dispatcher.add_handler(CommandHandler('add', add_item))
dispatcher.add_handler(CommandHandler('remove', remove_item))
dispatcher.add_handler(CommandHandler('confirm', confirm_order))

# start polling for updates
updater.start_polling()
updater.idle()
```
